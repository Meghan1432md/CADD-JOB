### Understanding Correlation

Correlation is a statistical measure that expresses the extent to which two variables are linearly related (meaning they change together at a constant rate).

*   **Positive Correlation (close to +1)**: Indicates that as one variable increases, the other variable also tends to increase.
*   **Negative Correlation (close to -1)**: Indicates that as one variable increases, the other variable tends to decrease.
*   **No Correlation (close to 0)**: Indicates little to no linear relationship between the two variables.

## Detect Outliers using IQR

### Subtask:
Calculate the Interquartile Range (IQR) for each numerical feature ('sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)') in the `df` DataFrame. Define upper and lower bounds to identify potential outliers for each feature.


**Reasoning**:
To detect outliers using the IQR method, I will first calculate the first quartile (Q1), third quartile (Q3), Interquartile Range (IQR), and then the lower and upper bounds for each numerical feature in the DataFrame.



```python
numerical_features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

outlier_bounds = {}

print("--- Outlier Bounds (IQR Method) ---")
for feature in numerical_features:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_bounds[feature] = {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'lower_bound': lower_bound,
        'upper_bound': upper_bound
    }
    print(f"\nFeature: {feature}")
    print(f"  Q1: {Q1:.2f}")
    print(f"  Q3: {Q3:.2f}")
    print(f"  IQR: {IQR:.2f}")
    print(f"  Lower Bound: {lower_bound:.2f}")
    print(f"  Upper Bound: {upper_bound:.2f}")
```

## Remove Detected Outliers

### Subtask:
Create a new DataFrame by filtering out the rows that were identified as outliers based on the IQR method from the 'sepal width (cm)' and other features. Display the shape of the original and new DataFrames to show the number of removed rows.


**Reasoning**:
To remove the identified outliers, I will create a new DataFrame `df_filtered` and sequentially apply the IQR-based filtering for each numerical feature, keeping only the data within the calculated lower and upper bounds. Finally, I will display the shapes of both the original and filtered DataFrames to show the impact of outlier removal.



```python
print("--- Removing Outliers ---")

df_filtered = df.copy()

initial_rows = df_filtered.shape[0]
print(f"Original DataFrame shape: {df.shape}")

for feature in numerical_features:
    bounds = outlier_bounds[feature]
    lower_bound = bounds['lower_bound']
    upper_bound = bounds['upper_bound']

    # Filter out outliers for the current feature
    df_filtered = df_filtered[(df_filtered[feature] >= lower_bound) & (df_filtered[feature] <= upper_bound)]

print(f"Filtered DataFrame shape: {df_filtered.shape}")
print(f"Number of rows removed: {initial_rows - df_filtered.shape[0]}")

print("--- First 5 rows of the filtered DataFrame ---")
print(df_filtered.head())
```
